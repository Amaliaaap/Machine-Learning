# 1. Develop a Line chart using the functionality of pandas to show how automobile sale fluctuate from year to year. Ticks on x axis with all the years to identify the years of recession annotation for at least two years of recession

```python
sales_df = df[['Recession', 'Automobile_Sales', 'Year']]

recession_years = sales_df[sales_df['Recession'] == 1]['Year'].tolist()

# Membuat plot
plt.figure(figsize=(20, 5))
plt.plot(sales_df['Year'], sales_df['Automobile_Sales'], marker='o', linestyle='-', color='b', label='Automobile_Sales')

# Menambahkan garis vertikal untuk menandakan tahun resesi
for year in recession_years:
    plt.axvline(x=year, color='r', linestyle='--', alpha=0.5)  # Garis merah untuk tahun resesi

plt.title('Fluktuasi Penjualan Mobil dari Tahun ke Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Penjualan')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

Berdasarkan line chart di atas, ada 5 kali resesi
1. tahun 1980, 1981, 1982, 1983

2. tahun 1990, 1991, 1992

3. tahun 2000, 2001, 2002

4. tahun 2008, 2009

5. tahun 2020

# 2. plot different line for categories of vehicle type and analyze the trend to answer the questions is there a noticeable different in sales trends between different vehicle types during recession periods?

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Create 'Recession' column
df['Recession'] = df['Date'].dt.year.isin(recession_years)

# Convert 'Automobile_Sales' to numeric, handling errors
df['Automobile_Sales'] = pd.to_numeric(df['Automobile_Sales'], errors='coerce')

# Group by year and vehicle type and calculate average sales
yearly_sales = df.groupby([df['Date'].dt.year, 'Vehicle_Type'])['Automobile_Sales'].mean().unstack()

# Create the plot
plt.figure(figsize=(15, 7))  # Adjust figure size as needed

# Plot each vehicle type
for vehicle_type in yearly_sales.columns:
    plt.plot(yearly_sales.index, yearly_sales[vehicle_type], marker='o', linestyle='-', label=vehicle_type)

# Highlight recession periods
for year in recession_years:
    plt.axvspan(year - 0.5, year + 0.5, color='red', alpha=0.3, label='Recession' if year == recession_years[0] else "")

plt.title('Automobile Sales Over Time by Vehicle Type with Recession Highlights')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.xticks(rotation=45)
plt.legend(title='Vehicle Type')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Dari data di atas, insight yang di dapat yaitu mengenai fluktuasi yang cukup besar dari tahun ke tahun untuk semua jenis kendaraan. Semua jenis kendaraan juga mengalami penurunan penjualan saat resesi, namun tingkat penuruna n dan pemulihan antar kendaraan berbeda.

# 4. Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.

**How did the GDP vary over time during recession and non-recession periods?**

```python
data = sales_df[['Recession', 'GDP', 'Year']] # Include 'Year' in the selection

# Assuming 'Recession' column has 1 for recession and 0 for non-recession
df_recession = data[data['Recession'] == 1]
df_non_recession = data[data['Recession'] == 0]
df = pd.DataFrame(data)
# Membuat subplot untuk membandingkan variasi GDP selama periode resesi dan non-resesi
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# Plot untuk resesi
axes[0].plot(df_recession['Year'], df_recession['GDP'], color='red', marker='o')
axes[0].set_title('GDP selama Periode Resesi')
axes[0].set_xlabel('Tahun')
axes[0].set_ylabel('GDP (triliun)')
axes[0].grid(True)

# Plot untuk non-resesi
axes[1].plot(df_non_recession['Year'], df_non_recession['GDP'], color='green', marker='o')
axes[1].set_title('GDP selama Periode Non-Resesi')
axes[1].set_xlabel('Tahun')
axes[1].set_ylabel('GDP (triliun)')
axes[1].grid(True)

# Menambahkan layout yang rapi
plt.tight_layout()
plt.show()
```

**Periode Resesi**

Grafik GDP pada periode resesi menunjukkan fluktuasi yang sangat tajam dan tidak stabil. Pada beberapa titik waktu, GDP mengalami penurunan yang sangat signifikan, bahkan mencapai titik terendah.

**Periode non-Resesi**

Grafik PDB non-resesi terlihat relatif lebih stabil dibandingkan saat resesi. Terlihat adanya tren pertumbuhan GDP dari waktu ke waktu selama periode ini. Meskipun terdapat fluktuasi kecil, namun arahnya cenderung meningkat.
