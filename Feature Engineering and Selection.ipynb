```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import numpy as np
np.set_printoptions(suppress=True)
```

**Scaling menggunakan StandardScaler**

```python
views = ['mileage', 'engine_size', 'num_owners']
scaler = StandardScaler()
df_encoded[views]=scaler.fit_transform(df_encoded[views])
df_encoded.head()
```

**Scaling menggunakan MinMaxScaler**

```python
views = ['mileage', 'engine_size', 'num_owners']
scaler = MinMaxScaler()
df_encoded[views]=scaler.fit_transform(df_encoded[views])
df_encoded.head()
```

**Scaling menggunakan RobustScaler**

```python
views = ['mileage', 'engine_size', 'num_owners']
scaler = RobustScaler()
df_encoded[views]=scaler.fit_transform(df_encoded[views])
df_encoded.head()
```

Perbedaan hasil antara StandardScaler, MinMaxScaler, dan RobustScaler adalah

1. StandarScaler mengubah data sehingga memiliki mean 0 dan standar deviasi 1 dan hasil scaling bisa negatif maupun positif.

2. MinMacScaler mengubah data ke dalan rentang 0 hingga 1 sehingga hasil scaling ada di antara o dan 1.

3. RobustScaler menggunakan median untuk scaling sehingga lebih tahan terhadap outlier, karena tidak menggunakan mean dan standar deviasi seperti StandardScaler.


Jadi, MinMaxScaler lebih baik digunakan jika data tidak berdistribusi normal dan data memiliki batasan nilai yang jelas. Misalnya, jika fitur memiliki rentang tetap, maka MinMaxScaler akan menjaga skala data tetap dalam batas yang wajar.

# **6.**

a. Buat fitur baru mileage_per_year dengan membagi mileage dengan car_age.

b. Mengapa mileage_per_year bisa menjadi fitur yang lebih informatif dibandingkan
mileage saja?

```python
views = ['mileage', 'engine_size', 'num_owners']
scaler = StandardScaler()
df_encoded[views]=scaler.fit_transform(df_encoded[views])
df_encoded.head()

df_encoded['mileage_per_year']=df_encoded['mileage']/df_encoded['car_age']
df_encoded.head()
```

```python
df_encoded['mileage_per_year']
```

Fitur mileage_per_year lebih informatif dibandingkan mileage saja karena fitur ini menggambarkan penggunaan mobil secara relatif. Jika Mileage hanya menunjukkan total jarak tempuh mobil tanpa memperhitungkan usia mobil, maka mileage_per_year mampu menunjukkan seberapa sering mobil digunakan per tahun sehingga bisa lebih relevan untuk menilai kondisi mobil.
